#### **ward data**

```python
#read in data
wards = pd.read_pickle('ward.p')

print(wards.shape)
wards.head()
```

#### **census data**

```python
census = pd.read_pickle('census.p')
print(census.shape)
census.head()
```

### **Merging the Datasets**

Dataset with be merged on the  `ward` column. The do share other columns that are the same `address` and `zip` but we are interested in merging on `ward` because we are interesting in data as it pertains to each ward

```python
wards_census = wards.merge(census, on='ward')
print(census.shape)
wards_census.head()
```

Now we have a dataframe containing 6 columns and 50 rows. Notice that there are 2 versions of the `address` and `zip` columns denoted with x and y suffixes. We can specify the suffixes that we want to use via the `suffixes` parameter. This will make it easier to identiy which column came from which of the original datasets

### **Merging the Datasets with Specifies Suffixes**

```python
wards_census = wards.merge(census, on='ward', suffixes=('_ward', '_cen')) #notice that the order matches the order that wards and census have 
                                                                          #ward - left, census - right --> _ward - left, _cen - right
wards_census.head()
```

### **One-to-Many Relationships**
* **one-to-one relationship** = every row in the left tabel is related to **only one** row in the right table
    * the wards and census data are examples of this 
* **one-to-many relationship** = every row in left table is related to **one or more** in right table
    * eg. there is a business license dataset which contains information about businesses within all of the different wards
    * there can be multiple businesses within each ward hence one-to-many relations

#### **merging `wards` and `business` datasets**

```python
business = pd.read_pickle('licenses.p')
business.head()
```

```python
wards_business = wards.merge(business, on='ward', suffixes=('_wards', '_biz'))
wards_business.head()
```

- similar to merging `wards` with `census` we used suffixes to denote the doubled columns and which dataframes they come from 
- in just showing the first 5 rows we can see that these rows repeat ward 1 and alderman Proco "Joe" Moreno but each row has a different business name and address (via `address_biz`)

#### **another one-to-many example: business with multiple owners**

Because business can have multiple owners the business owners and license dataframes are another example of a one-to-many relationship.

```python
biz_owners = pd.read_pickle('business_owners.p')
biz_owners.head()
```

- in the case of the `business` and `biz_owners` dataframes the column they have in common is `account`


```python
licenses_owners = business.merge(biz_owners, on='account')
licenses_owners.head()
```

- this shows another *one-to-many* relationship  but unless the merging of the other two dataframes there are no columns with suffixes because of all of the columns have unique names

### **Merging Multiple DataFrames**

Introducing 3 new datasets. `cal`, `ridership`, and`stations`. This data comes from Chicago public transit. We will use multiple merge to find the total number of rides provided to each passenger that passes through Wilson station during the weekdays in July. To do this we have to merge all 3 dataframes

```python
#read in new datasets
cal = pd.read_pickle('cta_calendar.p')
ridership = pd.read_pickle('cta_ridership.p')
stations = pd.read_pickle('stations.p')
```

```python
cal.head()
```

```python
ridership.head()
```

```python
stations.head()
```

- `cal` and `ridership` have year, month and day cols in common
- `stations` and `ridership` have station_id col in common

####  **merging dataframes**

```python
cal_ridership_stations = cal.merge(ridership, on=['year','month', 'day']).merge(stations, on='station_id')
cal_ridership_stations.head()
```

#### **finding total number of rides through wilson station, during the week, in july**

```python
#create filter of the criteria
filter = ((cal_ridership_stations['month']== 7)&
          (cal_ridership_stations['station_name'] == 'Wilson')&
          (cal_ridership_stations['day_type']=='Weekday'))
#use .loc to select rows that match filter and sum # of rides for all rows that this applies to 
answer = cal_ridership_stations.loc[filter, 'rides'].sum()
print(f'The answer is: {answer} rides were provided during the weekdays in july to passengers passing through Wilson station.')
```
